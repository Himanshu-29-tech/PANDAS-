## <center> Data Concatenation </center>
---
- Stacking dataFrmes On top of Each Other (rows) or side by side (Columns)
---


```python
import pandas as pd
```

```python
df1= pd.DataFrame({
    "id" : [1,2,3],
    "name" : ["Adam" , "Eve", "Bob"]
})

df2 = pd.DataFrame({
    "id" : [4,5,6],
    "name" : ["Charlie", "Faster", "Gilian"]
})

pd.concat([df1,df2]) # row

pd.concat([df1,df2], axis =1) #col
```
