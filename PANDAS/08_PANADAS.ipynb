## <center> Grouping & Aggregation of Data</center>
--- 
- groupby()
- meam/ sum / max / min / count 
- agg()
---

```python
import pandas as pd

df = pd.read_csv("employee_data.csv")
df2 =df.copy()


# groupby()

df.groupby("country")["income"].mean()
df.groupby("country")["income"].min()
df.groupby("country")["income"].max()


# mean/min/max 

df.groupby("country")["income"].agg(["mean", "min" , "max"])
df.groupby("country")["income"].aggregate(["mean", "min" , "max"])

# you can assign mean min and max to other values
df.groupby("country")["income"].agg(avg_salary="mean", min_salary = "min" ,max_salary = "max")

df.groupby("country").agg({
    "income" : "mean",
    "age" : "mean"
})

#rename to column

df.groupby("country").agg(
    max_salary = ("income" , "mean"),
    avg_age = ("age", "mean")
)
```

## 🔹 Grouping

Grouping means putting similar data into categories.

It is used to divide the dataset into smaller groups based on one or more columns.

---

## 🔹 Aggregation

Aggregation means applying functions to grouped data to get a single summarized value.

It helps in analyzing data easily.

### Common Aggregation Functions:

- `mean()` → Calculates average value  
- `sum()` → Calculates total value  
- `max()` → Finds highest value  
- `min()` → Finds lowest value  
- `count()` → Counts number of values
