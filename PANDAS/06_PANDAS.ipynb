## <center> Transforming Data(Feature engineering)</center>

- apply()
- map()
- assign()
- replace(old ,new)

```python
import pandas as pd
```

```python
# Feature engineering
df = pd.read_csv("raw_data.csv")
df2 = df.copy()


df2["tax"] = df2["Salary"].apply(lambda x : "20%" if x >= 50000 else "10%")


# suppose we wants to change Male = M, Female = F, Unknown + U

gender_map = {"Male" : "M", "Female" : "F", "Unknown" : "U"}
df2["gender"] = df2["gender"].map(gender_map)

#Assign
df2.assign(new_income = df2["Salary"]  * 1.1)

#Replace
df2["country"] = df2["country"].replace("USA","US")
df2
```
