##                               Data analaysis and manipulation tool 
---
- pandas are series datafarame
- series --> 1D labelled array
- Dataframe --> 2D labelled array

```python
import pandas as pd
```

```python
info = {
    "Name" : ["Rahul", "Harshita", "Shardha"],
    "CGPA" : [9.5, 7.5, 8.5]
}
df = pd.DataFrame(info)

print(df)
```

```python
s = pd.Series([1,2,3,4,5])
print(s)
print(type(s))

#index

print(s[0])
print(s[1])
print(s[2])
```

## Label

```python
age = pd.Series([23,24,25,26],index = ["Adam", "Eve", "Bob" , "Charlie"])
print(age)

print (age["Adam"])
print (age["Eve"])
print (age["Bob"])
print (age["Charlie"])

print(age.index)
```

##                                     series (properties)
---
- Homogenious
- Vectorized oprations
- handle Missing values With NaN
- mutable values,immutable size

---

```python
# vectorized operations
s1 =pd.Series([1,2,3,4,5])
s2 = pd.Series([10,20,30,40,50])

print(s1 + s2)

#mutable values
s1[0] = 100
print(s1)

#size immutable
changed_s1 =s1.drop(0)

print(s1)
print(changed_s1)
```

##                                                  <center>    DataFrame(2D Labeled Array) </center>

```python
# Dataframe

info = {
    "Name" : ["Admin", "Eye" , "Boss"],
    "Age"  : [23,24,25],
    "GPA"  : [9.5, 8.6,7.2]
}

df = pd.DataFrame(info)

print(df, type(df))

# 2D Array have two rows and columns

# for getings  rows
print(df.index)
# for getings columns
print(df.columns)
```

```python
# Methods of creating dataframe
df = pd.DataFrame([["Himanshu", 23], ["Rahul", 23], ["Elish", 35]], columns=["Name", "Age"])
print(df)
```

```python
import numpy as np 

np_arr = np.array([[1,2,3], [4,5,6]])

df = pd.DataFrame(np_arr, columns =["A", "B","C"])

print(df)
```
