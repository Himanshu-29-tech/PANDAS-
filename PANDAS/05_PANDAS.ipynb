## <center> Cleaning data </center>

```python
import pandas as pd
```

```python
# cleaning data - missing values

df = pd.read_csv("raw_data.csv")
df.isnull()
df.isna() # alias of isnull
df.isnull().sum()

df.dropna() # drop empty row
df.dropna(axis=1) # col drop

df.fillna(0)
df.fillna(100)

#ffill() -->> forward fill ---->copying the last value that are encounter from up to bottom
df.ffill()
df.bfill()
#bfill() -->> Backwarrd fill ----> copying the last value but from bottom to up
```
